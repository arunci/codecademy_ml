```python
import pandas as pd

ad_clicks = pd.read_csv("ad_clicks.csv")
ad_clicks.head()
```

```python
ad_clicks.groupby("utm_source").user_id.count().reset_index()
```

```python
ad_clicks["is_click"] = ~ad_clicks.ad_click_timestamp.isnull()
ad_clicks
```

```python
clicks_by_source = ad_clicks.groupby(["utm_source", "is_click"]).user_id.count().reset_index()
clicks_by_source
```

```python
clicks_pivot = clicks_by_source.pivot(
    columns="is_click",
    index="utm_source",
    values="user_id"
)
clicks_pivot
```

```python
clicks_pivot["percent_clicked"]=clicks_pivot[True] / (clicks_pivot[False] + clicks_pivot[True]) * 100
clicks_pivot
```

```python
clicks_by_experimental_group = ad_clicks.groupby(["experimental_group", "is_click"]).user_id.count().reset_index()
clicks_by_experimental_group
```

```python
clicks_by_experimental_group_pivot = clicks_by_experimental_group.pivot(
    columns="is_click",
    index="experimental_group",
    values="user_id"
).reset_index()
clicks_by_experimental_group_pivot
```
